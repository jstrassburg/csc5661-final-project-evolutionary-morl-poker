```python
class Base:
    def __init__(self):
        self.NUM_ACTIONS = 3

    def foo(self):
        return "bar"
    
    def print_num_actions(self):
        print(self.NUM_ACTIONS)
    
class Child(Base):
    def __init__(self):
        super().__init__()

    def bar(self):
        return self.foo()
    
    def bletch(self):
        return 4
    
class Child2(Child):
    def __init__(self):
        super().__init__()
        self.NUM_ACTIONS = 5

    def bar(self):
        return self.foo()
    
    def bletch(self):
        a = super().bletch()
        return a + 3
    
c = Child()
print(c.bar())
c.print_num_actions()
c2 = Child2()
print(c2.bar())
c2.print_num_actions()
print(c2.bletch())
```
